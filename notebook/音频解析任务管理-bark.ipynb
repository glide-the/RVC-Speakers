```python
from speakers.processors import get_processors,BarkProcessorData
from speakers.tasks import BarkVoiceFlowData, Runner
 


from collections import deque

vits_voice = get_processors('vits_processor')

rvc_processors = get_processors('rvc_processor')
 

SPEAKER = "zh_speaker_2"

#提示说话人应该说什么话，趋近于1，提示词特殊标记越明显
text_temp=1
#说话人紧张程度，趋近与1讲话越流畅，越接近与0，讲话人的声音听起来旁边有很多人，畏首畏尾
waveform_temp=0.9

bark_processor_data = BarkProcessorData(text='我是你的情感导师，青春的时间转瞬即逝，秋水于海又岂有一只浮游',
                          speaker_history_prompt=SPEAKER,
                          text_temp = text_temp,
                          waveform_temp = waveform_temp)


loaded_models=[
        '%s - %s' % (
            m['metadata'].get('source', 'Unknown'),
            m['metadata'].get('name')
        )
        for m in rvc_processors.loaded_models
]
 


#变调(整数, 半音数量, 升八度12降八度-12)
f0_up_key=-1
f0_method =["pm", "harvest", "crepe", "rmvpe"]

#检索特征占比
index_rate = 0.9
#>=3则使用对harvest音高识别的结果使用中值滤波，数值为滤波半径，使用可以削弱哑音
filter_radius=1
#输入源音量包络替换输出音量包络融合比例，越靠近1越使用输出包络
rms_mix_rate=1
#后处理重采样至最终采样率，0为不进行重采样
resample_rate=0

rvc_processor_data = RvcProcessorData(  
    model_index= 0,
    f0_up_key= f0_up_key,
    f0_method= f0_method[3],
    index_rate= index_rate,
    filter_radius= filter_radius,
    rms_mix_rate= rms_mix_rate,
    resample_sr= resample_rate 
)


# 创建一个 BarkVoiceFlowData 实例，并将 BarkProcessorData 实例作为参数传递
voice_flow_data = BarkVoiceFlowData(bark=bark_processor_data,
                               rvc=rvc_processor_data)

# 创建 Runner 实例并传递上面创建的 BarkVoiceFlowData 实例作为参数
runner = Runner(
    task_id="1",
    flow_data=voice_flow_data
)
 

```

```python
preprocess_object = bark_voice_task.preprocess_dict.get(bark_processor_data.type)
preprocess_object.match
 

import nltk  # we'll use this to split into sentences
import numpy as np
script = """我是你的情感导师，
青春的时间转瞬即逝，
秋水于海又岂有一只浮游""".replace("\n", "。").strip()
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[^，。！？]+[，。！？]?')
sentences = tokenizer.tokenize(script)
 
sentences
```

```python
 
rate,output_audio = await bark_voice_task.dispatch(runner=runner)
```
