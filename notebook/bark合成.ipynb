```python


import nltk  # we'll use this to split into sentences
import numpy as np
script = """
观众朋友大家好 [laughter]- 。
但她似乎在这场比赛中有些力不从心，局面对她来说并不太妙。 
""".replace("\n", ",").strip()
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(".*?[。！？]")
sentences = tokenizer.tokenize(script)
 
sentences
```

```python
from speakers.processors import  BarkProcessorData

# Set up sample rate
SAMPLE_RATE = 24_000 
SPEAKER = "zh_speaker_2"
silence = np.zeros(int(0.25 * SAMPLE_RATE))  # quarter second of silence

pieces = []

#提示说话人应该说什么话，趋近于1，提示词特殊标记越明显
text_temp=1
#说话人紧张程度，趋近与1讲话越流畅，越接近与0，讲话人的声音听起来旁边有很多人，畏首畏尾
waveform_temp=0.9

for sentence in sentences:
    
    data = BarkProcessorData(text=sentence,
                      speaker_history_prompt=SPEAKER,
                      text_temp = text_temp,
                      waveform_temp = waveform_temp)
    audio_array = bark_voice(data=data)
    pieces += [audio_array, silence.copy()]
 
```

```python
Audio(np.concatenate(pieces), rate=SAMPLE_RATE)
```
