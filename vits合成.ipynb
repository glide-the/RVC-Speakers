```python
from vits import VitsToVoice
import sys
import pandas as pd
from tqdm import tqdm
```

```python

abs_path = sys.path[0] 

vits_model=abs_path+'/vits/model/G_953000.pth'
voice_config_file = abs_path+'/vits/model/config.json'
vits_voice = VitsToVoice(vits_model=vits_model,voice_config_file=voice_config_file)
```

```python

speakers_table = pd.DataFrame(vits_voice.speakers, columns=["speaker"])     
speakers_table
```

```python
 
language_table = pd.DataFrame(vits_voice.lang, columns=["language"])     
language_table
```

```python

input_text = '我是你的情感导师，青春的时间转瞬即逝，秋水于海又岂有一只浮游'
lang = vits_voice.lang.index('简体中文') 
speaker_id = vits_voice.speakers.index('周良')
print(speaker_id)
print(lang)
#param noise_scale: noise_scale(控制感情变化程度)
noise_scale = 1
 
#param noise_scale_w: noise_scale_w(控制音素发音长度)
noise_scale_w= 1.0
# (minimum=0.1, maximum=5, value=1, step=0.1,  label='速度 Speed') 控制整体语速
speed = 1


output_audio = vits_voice.vits_func(text=input_text,language=lang,speaker_id=speaker_id, 
                    noise_scale=noise_scale, noise_scale_w=noise_scale_w, speed=1)



```

```python
output_audio
```

```python
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio
SAMPLE_RATE = 22050 
# save audio to disk
write_wav(f"test.wav", SAMPLE_RATE, output_audio) 

# play text in notebook
Audio(output_audio, rate=SAMPLE_RATE)
```
